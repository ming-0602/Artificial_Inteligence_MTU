```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn import svm
```

```python
kfld = KFold(n_splits=5, shuffle=True)
```

```python
iris_data = load_iris()
```

```python
iris_data
```

```python
X = iris_data['data']
Y = iris_data['target']
```

```python
clf = svm.SVC()
```

```python
clf.fit(X, Y)
```

```python
kfld = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# kfld.split(iris_data['data'])
```

```python
for training_indices, test_indices in kfld.split(iris_data['data']):
    print(training_indices, test_indices)
    X_train = iris_data['data'][training_indices]
    Y_train = iris_data['target'][training_indices]
    X_test = iris_data['data'][test_indices]
    Y_test = iris_data['target'][test_indices]
    clf.fit(X_train, Y_train)
    print(clf.score(X_test, Y_test))
```
